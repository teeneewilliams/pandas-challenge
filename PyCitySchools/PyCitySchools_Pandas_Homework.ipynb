```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# calculate the total number of schools by counting unique schools 
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
# calculate the total number of students
# find the number of unique values over the index axis in the dataframe
num_students = school_data_complete.nunique(axis=1)
print(len(num_students))
```

```python
# calculate total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# calculate average math score
avg_rd_score = round(school_data_complete["reading_score"].mean(),2)
avg_rd_score
```

```python
# calculate average reading score
avg_mth_score = round(school_data_complete["math_score"].mean(),2)
avg_mth_score
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
passing_math = len(school_data_complete.loc[school_data_complete["math_score"] >= 70,:])
percent_math = (passing_math/len(num_students))*100
percent_math
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
passing_read = len(school_data_complete.loc[school_data_complete["reading_score"] >=70,:])
percent_read = (passing_read/len(num_students))*100
percent_read
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
passing_students = (((passing_math + passing_read)/len(num_students))/2)*100
passing_students
```

```python
# Create a dataframe to hold the above results
summary_df = pd.DataFrame({"Total Schools":[school_count],
                           "Total Students":["{:,}".format(len(num_students))],
                           "Total Budget": ["${:,.2f}".format(total_budget)],
                           "Avg Reading Score": [avg_rd_score],
                           "Avg Math Score": [avg_mth_score],
                           "% Passing Math": ["{:,.2f}%".format(percent_math)],
                           "% Passing Reading": ["{:,.2f}%".format(percent_read)],
                           "% Overall Passing": ["{:,.2f}%".format(passing_students)]})
summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Sort by school name
sorted_schools_df = school_data_complete.sort_values("school_name", ascending=True)
sorted_schools_df
```

```python
# School Type by school name using the school df
school_type = school_data.set_index(["school_name"])["type"]
```

```python
# Total Students using the complete sorted df
total_students = sorted_schools_df["school_name"].value_counts()    
```

```python
# School Budget from the school df
school_budget = school_data.set_index(["school_name"])["budget"]
```

```python
# Calculate budget Per Student
budget_per_student = school_budget / total_students
```

```python
# Calculate average Math Scores from groupby sorted schools full df
per_school_math_average = sorted_schools_df.groupby(["school_name"]).mean()["math_score"]
```

```python
# Calculate average Reading Scores from groupby sorted schools full df
per_school_reading_average = sorted_schools_df.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Percent Passing using sorted full df
school_percent_passing_math = ((sorted_schools_df[(sorted_schools_df["math_score"] >= 70)].groupby(["school_name"]).count()["student_name"]) / total_students) * 100
school_percent_passing_reading = ((sorted_schools_df[(sorted_schools_df["reading_score"] >= 70)].groupby(["school_name"]).count()["student_name"]) / total_students) * 100
```

```python
# Calculate overall passing rate using per passing for math and reading calculated
overall_passing_rate = (school_percent_passing_math + school_percent_passing_reading) / 2
                          
```

```python
# Create the dataframe and read in the series
school_summary_df = pd.DataFrame({"School Type": school_type, 
                                  "Total Students": total_students,
                                  "Total School Budget": school_budget, 
                                  "Per Student Budget": budget_per_student, 
                                  "Average Math Score": per_school_math_average,
                                 "Average Reading Score": per_school_reading_average, 
                                  "% Passing Math": school_percent_passing_math,
                                  "% Passing Reading": school_percent_passing_reading,
                                  "% Overall Passing": overall_passing_rate})


# Formatting
school_summary_df["Total Students"] = school_summary_df["Total Students"].map("{:,.0f}".format)
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df["Average Math Score"] = school_summary_df["Average Math Score"].map("{:,.2f}".format)
school_summary_df["Average Reading Score"] = school_summary_df["Average Reading Score"].map("{:,.2f}".format)
school_summary_df["% Passing Math"] = school_summary_df["% Passing Math"].map("{:,.2f}".format)
school_summary_df["% Passing Reading"] = school_summary_df["% Passing Reading"].map("{:,.2f}".format)
school_summary_df["% Overall Passing"] = school_summary_df["% Overall Passing"].map("{:,.2f}".format)
school_summary_df
```
